```python
import requests
import regex as re
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36"
}
url="https://www.bikewale.com/"
# https://www.bikewale.com/royalenfield-bikes/
```

```python
response = requests.get(url, headers=headers)
print(response)
```

```python
Rating=[]
NumberOfRating=[]
Brand=[]
CC=[]
Milage=[]
Power=[]
Weight=[]
Price=[]
```

```python
soup=BeautifulSoup(response.text, "html.parser")
soup1=soup.find_all("div",class_="o-hl o-iT o-o")
temp=soup1[0].find_all("a",class_="o-d o-os")
# print(temp.get("href"))
c=0
count=0
for j in temp:
    link=f"https://www.bikewale.com{j.get("href")}"
    res=requests.get(link, headers=headers)
    soup=BeautifulSoup(res.text, "html.parser")
    soup1=soup.find_all("div",class_="o-C o-cF o-c9 o-bT o-cp o-f")
    for i in soup1:
        var=i.text
        if re.findall("^\d",var):
            tempp=i.find("a",class_="o-f o-aF o-jJ o-eQ")
            list1=tempp.get("href").split("/")
            link2=link+list1[-2]
            # print(link2)
            res1=requests.get(link2, headers=headers)
            soup2=BeautifulSoup(res1.text, "html.parser")
            var1=soup2.find_all("td",class_="o-js o-j1 o-jh o-jJ")
            if len(var1)>=2:
                mcount=0
                for k in var1:
                    if "kmpl" in k.text:
                        Milage.append(re.findall("(\d?\d?\d?\.?\d+)\skmpl",k.text)[0])
                        mcount=1
                if not mcount:
                    Milage.append(np.nan)
            else:
                Milage.append(np.nan)
            var1=i.find("div",class_="o-o FTJfQA o-iT o-ei").text
            var2=i.find("div",class_="o-jK o-ei o-fu").text
            var3=i.find("div",class_="o-eg o-fm o-m5 o-mu o-mL o-na o-fzqLoc o-bv o-h o-bc").text
            var4=i.find("h3",class_="o-j4 o-jq o-hM o-c4 o-jK").text
            temp1=re.findall("(\d.+)\scc",var1)[0]if re.findall("(\d.+)\scc",var1) else np.nan
            CC.append(temp1)
            # temp2=re.findall("\|(\d.+)\skmpl",var1)[0] if re.findall("\|(\d.+)\skmpl",var1) else np.nan
            # Milage.append(temp2)
            temp3= re.findall("(\d?\d?\d?\.?\d+)\sbhp",var1)[0] if re.findall("\d?\d?\d?\.?\d+\sbhp",var1) else np.nan
            Power.append(temp3)
            temp4=re.findall("(\d?\d?\d?\.?\d+)\skg",var1)[0] if re.findall("\d?\d?\d?\.?\d+\skg",var1) else np.nan
            Weight.append(temp4)
            temp5=re.findall("(\w.+)Avg",var2)[0] if re.findall("(\w.+)Avg",var2) else np.nan
            Price.append(temp5)
            temp6=re.findall("(.+)\/",var3)[0] if re.findall("(.+)\/",var3) else np.nan
            Rating.append(temp6)
            temp7=re.findall("\/(.+\d)\sRating",var3)[0] if re.findall("\/(.+\d)\sRating",var3) else np.nan
            NumberOfRating.append(temp7)
            Brand.append(var4)
    # c+=1
    # if c==3:
    #     break
    
print(count)
print(len(Rating))
```
