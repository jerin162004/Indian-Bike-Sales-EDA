```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("bike_data_set1.csv")
```

```python
df.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
df.disc
```

```python

df.head(60)
```


## univariate 

```python
# HISTOGRAM (Price)
plt.figure(figsize=(8,5),facecolor="#dff2e1")
sns.histplot(df['Price'], bins=20, color='skyblue')
plt.xlabel("Price ()")
plt.ylabel("Frequency")
plt.title("Price Distribution (Histogram)")
plt.show()
```

```python
# BOXPLOT (Mileage)
plt.figure(figsize=(8,5),facecolor="#dff2e1")
sns.boxplot(x=df['Mileage'], color='lightgreen')
plt.xlabel("Mileage (kmpl)")
plt.title("Mileage Boxplot")
plt.show()
```

```python
# HISTOGRAM (Engine CC)
plt.figure(figsize=(8,5),facecolor="#dff2e1")
sns.histplot(df['CC'], bins=20, color='orange')
plt.xlabel("Engine CC")
plt.ylabel("Count")
plt.title("Engine CC Distribution")
plt.show()
```

```python
# BOXPLOT (Power)
plt.figure(figsize=(8,5),facecolor="#dff2e1")
sns.boxplot(x=df['Power'], color='pink')
plt.xlabel("Power (bhp)")
plt.title("Power Boxplot")
plt.show()
```

```python
# BAR CHART (Brand Count)
plt.figure(figsize=(12,6),facecolor="#dff2e1")
df['Brand_Name'].value_counts().plot(kind='bar', color='purple')
plt.xlabel("Brand_Name")
plt.ylabel("Number of Bikes")
plt.title("Brand Frequency (Bar Chart)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sizes = df["Brand_Name"].value_counts()
plt.pie(sizes, labels=sizes.index, autopct="%1.1f%%", wedgeprops={'width':0.4})
plt.title("Donut Chart: Models per Brand")
plt.show()
```

```python
#SCATTER PLOT  CC vs Price
plt.figure(figsize=(8,5))
plt.scatter(df['CC'], df['Price'], alpha=0.6)
plt.xlabel("Engine CC")
plt.ylabel("Price : Eg(1*10=10 Lakhs)")
plt.title("CC vs Price")
plt.show()
```

```python
# SCATTER PLOT  Power vs Mileage
plt.figure(figsize=(8,5))
plt.scatter(df['Power'], df['Mileage'], color='green', alpha=0.6)
plt.xlabel("Power (bhp)")
plt.ylabel("Mileage (kmpl)")
plt.title("Power vs Mileage")
plt.show()
```

```python
#LINE PLOT  CC vs Power
plt.figure(figsize=(8,5))
plt.plot(df['CC'], df['Power'], linestyle='-', marker='o', alpha=0.5)
plt.xlabel("Engine CC")
plt.ylabel("Power (bhp)")
plt.title("CC vs Power (Line Plot)")
plt.show()
```

```python
#SCATTER PLOT  Power vs Price
plt.figure(figsize=(8,5))
plt.scatter(df['Power'], df['Price'], alpha=0.6)

plt.xlabel("Power (bhp)")
plt.ylabel("Price : Eg(1*10=10 Lakhs)")
plt.title("Power vs Price")
plt.show()
```

```python
# bar chart - Calculate average mileage per brand
brand_mileage = df.groupby("Brand_Name")["Mileage"].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(12,6))
brand_mileage.plot(kind='bar', color='skyblue')

plt.xlabel("Brand_Name", fontsize=12)
plt.ylabel("Average Mileage (kmpl)", fontsize=12)
plt.title("Average Mileage by Brand", fontsize=14)
plt.show()
```

```python
#BOXPLOT  Brand vs Price (Top 10 Brands)
plt.figure(figsize=(12,6))
top_brands = df['Brand_Name'].value_counts().head(10).index
subset = df[df['Brand_Name'].isin(top_brands)]

subset.boxplot(column='Price', by='Brand_Name', figsize=(12,6))
plt.title("Brand_Name vs Price ()")
plt.suptitle("")  # remove default title
plt.xticks(rotation=45)
plt.show()
```

```python
# BAR PLOT  Average Mileage by Brand (Top 10 Brands)
plt.figure(figsize=(12,6))
subset.groupby('Brand_Name')['Mileage'].mean().head(10).plot(kind='bar', color='orange')
plt.xlabel("Brand_Name")
plt.ylabel("Average Mileage (kmpl)")
plt.title("Average Mileage by Top 10 Brands")
plt.xticks(rotation=45)
plt.show()
```

```python
cols = ['Price', 'CC', 'Mileage', 'Power', 'Weight', 'Rating']
plt.figure(figsize=(9,6))
sns.heatmap(df[cols].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection="3d")

sc = ax.scatter(
    df["CC"], df["Power"], df["Price"],
    c=df["Mileage"], cmap="viridis", s=40
)

ax.set_xlabel("Engine CC")
ax.set_ylabel("Power (bhp)")
ax.set_zlabel("Price ()")
plt.title("3D Scatter: CC vs Power vs Price (Color = Mileage)")

fig.colorbar(sc, label="Mileage (kmpl)")
plt.show()
```
